#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO
import pandas as pd

address1 = "Valdemorillo"
address2 = "Galapagar"

ds['pps'] = ds['price'] / ds['surface']

ds_address1 = ds[ds['address'] == address1]
ds_address2 = ds[ds['address'] == address2]

average_pps1 = ds_address1['pps'].mean() if not ds_address1.empty else None
average_pps2 = ds_address2['pps'].mean() if not ds_address2.empty else None

if average_pps1 is not None:
    print(f"Average price per square meter for {address1}: ${average_pps1:,.2f}")
else:
    print(f"No data found for {address1}")

if average_pps2 is not None:
    print(f"Average price per square meter for {address2}: ${average_pps2:,.2f}")
else:
    print(f"No data found for {address2}")

if average_pps1 is not None and average_pps2 is not None:
    if average_pps1 == average_pps2:
        print("\nThe average prices per square meter for both addresses are the same.")
    else:
        print("\nThe average prices per square meter for the two addresses are different.")
```

#### Exercise 11. Analyze the relation between the surface and the price of the houses (★★☆)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt

normalized_prices = {}

for population in south_belt_df['level5'].unique():

    population_df = south_belt_df[south_belt_df['level5'] == population].copy()
    
    min_price = population_df['price'].min()
    max_price = population_df['price'].max()
    
    population_df.loc[:, 'normalized_price'] = (population_df['price'] - min_price) / (max_price - min_price)
    
    normalized_prices[population] = population_df['normalized_price']

plt.figure(figsize=(10, 6))

for population, prices in normalized_prices.items():
    plt.hist(prices, bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices Distribution for South Belt of Madrid')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
from ipyleaflet import Map, Marker, MarkerCluster, Icon
import pandas as pd
from ipyleaflet import basemaps

map = Map(center=(40.4168, -3.7038), zoom=10, min_zoom=1, max_zoom=20, 
          basemap=basemaps.OpenStreetMap.Mapnik)  

population_colors = {
    "Fuenlabrada": "blue",
    "Leganés": "green",
    "Getafe": "red",
    "Alcorcón": "purple"
}

markers = []

for _, row in south_belt_df.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    population = row['level5']

    color = population_colors.get(population, "gray")  
    
    icon = Icon(color=color, icon_size=(25, 25))  
    
    marker = Marker(location=(latitude, longitude), draggable=False, 
                    title=population, icon=icon)
    
    markers.append(marker)

marker_cluster = MarkerCluster(markers=markers)
map.add_layer(marker_cluster)

map
```
